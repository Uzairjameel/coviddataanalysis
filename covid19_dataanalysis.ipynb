# Create a scatter plot comparing total cases and total deaths.

```python

# 📌 Sample Data
x = [10, 20, 30, 40, 50]
y = [3, 7, 2, 8, 5]
labels = ['A', 'B', 'C', 'D', 'E']  # ہر پوائنٹ کا نام

# 📌 Scatter Plot بنائیں گے
plt.scatter(x, y, color='blue')

# 📌 Labels پوائنٹس کے ساتھ لگائیں گے
for i in range(len(x)):
    plt.text(x[i], y[i], labels[i], fontsize=12, ha='right', va='bottom')

# 📌 Labels اور Title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Labels')

# 📌 Show Plot
plt.show()
```

```python


# 📌 Sample Data: Total Cases & Total Deaths
total_cases = [10000, 50000, 200000, 500000, 1000000, 2000000, 5000000]
total_deaths = [50, 500, 3000, 10000, 25000, 60000, 150000]

# 📌 Scatter Plot بنائیں گے
plt.figure(figsize=(8, 5))  # گراف کا سائز بڑا کریں گے
plt.scatter(total_cases, total_deaths, color='red', alpha=0.6, s=100)

# 📌 Labels اور Title
plt.xlabel('Total Cases')  # X-axis لیبل
plt.ylabel('Total Deaths')  # Y-axis لیبل
plt.title('Scatter Plot: Total Cases vs Total Deaths')  # ٹائٹل

# 📌 Log Scale لگائیں تاکہ ڈیٹا کو بہتر دیکھا جا سکے
plt.xscale('log')
plt.yscale('log')

# 📌 Show Plot
plt.show()
```
